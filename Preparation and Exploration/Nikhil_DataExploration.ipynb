```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_palette("pastel")
```

```python
student=pd.read_csv("student-mat-v1.csv")
```

```python
student
```

```python
fs_gt3=student['famsize'].loc[student['famsize']=='GT3']
fs_le3=student['famsize'].loc[student['famsize']=='LE3']
labels = 'Greater than 3', 'Less than 3'
levels = [fs_gt3.size, fs_le3.size]
fig,p = plt.subplots()
p.pie(levels, autopct='%1.1f%%')
p.set_title("% of students with family size")
plt.legend(labels)
plt.show()
```

```python
fs_gt3=student.loc[student['famsize']=='GT3',['y_binary']]
fs_gt3_p=fs_gt3['y_binary'].loc[fs_gt3['y_binary']==1]
fs_gt3_f=fs_gt3['y_binary'].loc[fs_gt3['y_binary']==0]
labels = 'Pass', 'Fail'
levels = [fs_gt3_p.size, fs_gt3_f.size]
fig,p = plt.subplots()
p.pie(levels, autopct='%1.1f%%')
p.set_title("Pass% vs Fail% with students having family size>3")
plt.legend(labels)
plt.show()
```

```python
fs_LE3=student.loc[student['famsize']=='LE3',['y_binary']]
fs_LE3_p=fs_LE3['y_binary'].loc[fs_LE3['y_binary']==1]
fs_LE3_f=fs_LE3['y_binary'].loc[fs_LE3['y_binary']==0]
labels = 'Pass', 'Fail'
levels = [fs_LE3_p.size, fs_LE3_f.size]
fig,p = plt.subplots()
p.pie(levels, autopct='%1.1f%%')
p.set_title("Pass% vs Fail% with students having family size<3")
plt.legend(labels)
plt.show()
```

From the above pie charts, we can infer that family size doesn't really effect the student's grade.
